## Speedup analysis in supervised binary classification problems

```python
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
```

### Hyper-Stacked vs Linear speedup

```python
FLIGHT = [[1, 1], [8, 1.815], [24, 1.899]]
SUSY = [[1, 1], [8, 2.934], [24, 2.944]]
HEPMASS = [[1, 1], [8, 3.138], [24, 4.413]]
HIGGS = [[1, 1], [8, 3.970], [24, 4.726]]
ideal = [[1, 1], [8, 8], [24, 24]]

x1,y1 = zip(*FLIGHT)
x2,y2 = zip(*SUSY)
x3,y3 = zip(*HEPMASS)
x4,y4 = zip(*HIGGS)
x5,y5 = zip(*ideal)

plt.plot(x1,y1, 'o-', label='FLIGHT')
plt.plot(x2,y2, 'o-', label='SUSY')
plt.plot(x3,y3, 'o-', label='HEPMASS')
plt.plot(x4,y4, 'o-', label='HIGGS')
plt.plot(x5,y5, 'k--',  linestyle = 'dotted',  label='Ideal')

plt.legend()
plt.xlabel("Number of cores")
plt.ylabel("Speedup")
plt.savefig('Speed_up1a.png', bbox_inches='tight', dpi=600)
plt.show()
```

### Training time comparison between clusters in HIGGS dataset (Speedup experiment with HIGGS dataset using different data sizes)

```python
x = np.arange(3)
HIGGS_100 = [109998.22, 20659.79, 10773.94]
HIGGS_33 = [35474.87, 8936.84, 7506.33]

width = 0.2
  
# plot data in grouped manner of bar type
plt.bar(x-0.2, HIGGS_100, width, color='green')
plt.bar(x, HIGGS_33, width, color='orange')

plt.xticks(x, ['Cluster 1', 'Cluster 2', 'Cluster 3'])
plt.legend(["HIGGS_100", "HIGGS_33"])
plt.xlabel("Clusters")
plt.ylabel("Training time (s)")
plt.savefig('Speedup_3a.png', bbox_inches='tight', dpi=600)
plt.show()
```
